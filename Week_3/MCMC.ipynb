```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, beta
```

# Exercise 1:

```python
a = 5
b = 17


def likelihood(n_heads, n_coins, p):
    return binom.pmf(n_heads, n_coins, p)

n_coins = 100
n_heads = 66
p = np.linspace(0, 1, 100)
prior = beta.pdf(p, a, b)
lh = likelihood(n_heads, n_coins, p)
posterior = prior * lh

lh_text = r'$LH = \frac{n!}{k!(n-k)!}p^k(1-p)^{n-k} = \frac{100!}{66!(100-66)!}p^{66}(1-p)^{100-66}$'


fig, ax = plt.subplots(3, 1, figsize=(10, 6), sharex=True)
ax[0].plot(p, prior)
ax[0].set_title('Prior')
ax[1].plot(p, lh)
ax[1].set_title('Likelihood')
ax[1].text(0.05, 0.75, lh_text, fontsize=12, color='black', transform=ax[1].transAxes)
ax[2].plot(p, posterior)
ax[2].set_title('Posterior')
ax[2].set_xlabel('p')

for a in ax.flatten():
    a.set_facecolor('lightgrey')
    a.grid(color='white', linestyle='--', linewidth=0.5)
plt.tight_layout()



```

# Exercise 2: 
MCMC

```python
start_1 = 100
start_2 = -27

x1 = [start_1]
x2 = [start_2]

N_steps = 100
n = 0

while n <= N_steps:
    x1_new = np.random.normal(0.5*x1[-1], 1)
    x2_new = np.random.normal(0.5*x2[-1], 1)
    x1.append(x1_new)
    x2.append(x2_new)

    n += 1
n_space = np.linspace(0, N_steps, N_steps)
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(x1, label='x1')
ax.plot(x2, label='x2')
ax.set_facecolor('lightgrey')
ax.grid(color='white', linestyle='--', linewidth=0.5)



```
